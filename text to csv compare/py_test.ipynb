```python
# load dataset CSV
working_dir = os.getcwd()
print(working_dir)
data_dir = os.path.join(r'C:\Users\jaske\OneDrive\Desktop\New folder','file.csv')
csv_data = pd.read_csv(data_dir)
# select only needed data - name & amount columns
csv_sorted = csv_data[['Name','Amount']].sort_values(by='Amount',ascending=True).reset_index(drop=True)

weapons_list = []
with open(r'C:\Users\jaske\OneDrive\Desktop\New folder\wepnames.txt','r') as f:
    lines = f.readlines()
    file_content = ''.join(lines)
#join lines and append to the weapon_list
for line in file_content.splitlines(''):
    weapons_list.append(line.strip())
    
items = pd.Series(weapons_list).str.strip()

#merge 2 left (CSV) and right (weapons)
weapons_df = pd.DataFrame(columns=['Weapons'])
weapons_df['Weapons'] = items
#print(weapons_df)

merged_df = pd.merge(csv_sorted,weapons_df,how='right',left_on='Name',right_on='Weapons').sort_values(ascending=True,by='Amount')
print(merged_df)
```

```python
merged_df = pd.merge(csv_sorted,weapons_df,how='right',left_on='Name',right_on='Weapons').sort_values(ascending=True,by='Amount').reset_index(drop=True)
display(merged_df)
sumnans = merged_df['Name'].isna().sum()
print(sumnans)
```

```python
results = merged_df['Weapons'].where(merged_df['Name'].isna())
merged_df['results'] = results
results = results.dropna().reset_index(drop=True)
results_str = results.to_string()
#got it
```

```python
print(type(results_str))
```

```python
with open('results.txt','w') as f1:
    f1.write(f'Items that are not in the CSV file. Total:  {results.value_counts().sum()} items. \n')
    f1.write(f'\n{results_str}\n')
    f1.write(f'--'*50)
   
```
